```python
def get_generation(input, model, tokenizer):
  print(len(input))
  input_ids = tokenizer(input, max_length=1024, truncation = True, return_tensors="pt").input_ids
  input_ids = input_ids.to(model.device)
  print(f"Token count: {input_ids.shape[1]}")

  output_ids = model.generate(input_ids, max_length=64, num_beams=5,
                                    early_stopping=True,
                                    no_repeat_ngram_size=2)
  decoded = tokenizer.decode(output_ids[0], skip_special_tokens=True)
  return decoded
```

### Preparing DataLoader


```python
ds = ds.filter(lambda x: x['description'] is not None and len(x['description']) < (1024 - prompt_length) and x['tags'] is not None and x['rating'] is not None)
```

```python
print(ds)
```

```python
train_data = ds['train']
test_data = ds['test']
```

```python
from torch.utils.data import Dataset, DataLoader

class CodeForcesDataset(Dataset):
  def __init__(self, data, tokenizer, prompt, max_input_len = 1024, max_target_len = 128):
    self.data = data
    self.tokenizer = tokenizer
    self.prompt = prompt
    self.max_input_len = max_input_len
    self.max_target_len = max_target_len

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    row = self.data[idx]
    problem_description = get_problem_description(self.prompt, row)
    tags = row['tags']
    difficulty = max(0, min(1, ((row['rating'] or 0) - 800) / (3500 - 800)))
    target = format_target(tags, difficulty)

    input_enc = self.tokenizer(
        problem_description,
        padding = "max_length",
        truncation = True,
        max_length = self.max_input_len,
        return_tensors="pt")

    target_enc = self.tokenizer(
        target,
        padding = "max_length",
        truncation = True,
        max_length = self.max_target_len,
        return_tensors="pt")

    labels = target_enc['input_ids'].squeeze()
    labels[labels == self.tokenizer.pad_token_id] = -100

    return {
      'input_ids': input_enc['input_ids'].squeeze(),
      'attention_mask': input_enc['attention_mask'].squeeze(),
      'labels': labels
    }
```

```python
train_dataset = CodeForcesDataset(train_data, tokenizer, prompt2)
test_dataset = CodeForcesDataset(test_data, tokenizer, prompt2)
```

### Model

```python
from peft import get_peft_model, LoraConfig, TaskType

peft_config = LoraConfig(
    task_type=TaskType.SEQ_2_SEQ_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.1,
    bias="none"
)
```

```python
peft_model = get_peft_model(model, peft_config)
peft_model.print_trainable_parameters()
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./ts-small-tag-encoder",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=5,
    learning_rate=3e-4,
    eval_strategy="steps",
    eval_steps=500,
    load_best_model_at_end=True,
    fp16=True,
    report_to = "none"
)

trainer = Trainer(
    model=peft_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)
```

```python
trainer.train()
```
